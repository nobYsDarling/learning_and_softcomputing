## 1 Naïve Bayes und die bedingte Wahrscheinlichkeit

In diesem Kapitel wird die Theorie zu dem Klassifizierungsverfahren des Naïve Bayes vorgestellt. Zu dem Verfahren existieren für die Implementierung verschiedenen Ausprägungen, welche die Wahrscheinlichkeiten der Merkmalabhängigkeiten unterschiedlich berechnen. 

### 1.1 Klassifizierung mit der Bayeschen Entscheidungstheorie

Die Basis des Maschinellen Lernens mittels des Naïve Bayes stellen Theoreme der Wahrscheinlichkeitstheorie dar. Der Naïve Bayes bildet eine Untermenge der Bayeschen Entscheidungstheorie. Im Folgenden sei eine Menge von Daten vorhanden, die sich in zwei verschiedene Klassen unterteilen lassen. Jeder Datensatz kann durch ein Tupel $(x,y)$ repräsentiert werden; dabei sind x und y die Parameter, die darüber entscheiden, ob ein Datensatz eher der einen oder eher der anderen Klasse zuzuordnen ist. Dafür existieren die zwei Funktionen $P_1(x,y)$ und $P_2(x,y)$. Dabei gelten die folgenden Regeln:

Ist $P_1(x,y) > P_2(x,y)$, so ist der Datensatz der Klasse 1 zuzuordnen.<br />
Ist $P_2(x,y) > P_1(x,y)$, so ist der Datensatz der Klasse 2 zuzuordnen.

Zusammengefasst bedeutet das, dass im Rahmen der bayeschen Entscheidungstheorie die Entscheidung gewählt wird, deren Wahrscheinlichkeit am größten ist.

### 1.2 Bedingte Wahrscheinlichkeit

Es soll im Folgenden bedingte Wahrscheinlichkeiten näher betrachtet werden, um den gezeigten, sehr grundlegenden, Ansatz zu erweitern. Zur Verdeutlichung wird ein Behälter mit sieben Steinen betrachtet, von denen drei grau und vier schwarz sind. Es ist ohne Weiteres ersichtlich, dass die Wahrscheinlichkeit, einen grauen Stein zu ziehen $\frac{3}{7}$ und einen schwarzen Stein zu ziehen $\frac{4}{7}$ beträgt. Dies kann notiert werden als $P(grau) = \frac{3}{7}$ bzw. $P(schwarz) = \frac{4}{7}$. Zur Berechnung des Ergebnisses wurde lediglich die Anzahl der zu betrachtenden Steine ins Verhältnis zur Anzahl der Gesamtmenge gesetzt.

Interessanter wird die Fragestellung, sobald man die sieben Steine zufällig auf zwei verschiedene Behälter verteilt. Soll nun die Wahrscheinlichkeit für das Ziehen eines grauen oder schwarzen Steins bestimmt werden, so würde die Kenntnis des Behälters, aus dem gezogen wird, das bisherige Ergebnis verändern.
Im Folgenden soll davon ausgegangen werden, dass sich in Behälter A zwei graue und zwei schwarze Steine befinden, wohingegen sich der Rest, also ein grauer und zwei schwarze Steine, in Behälter B befinden.
Es soll nun also bestimmt werden, wie hoch die Wahrscheinlichkeit ist, einen grauen Stein zu ziehen, wenn bekannt ist, dass es sich um den Behälter B handelt. Formal kann dies geschrieben werden als $P(\,grau \mid Behälter_B\,)$ und kann gelesen werden als „die Wahrscheinlichkeit, dass es sich um einen grauen Stein handelt, unter der Voraussetzung, dass Behälter B ausgewählt wurde“. In dem genannten Beispiel ist deutlich ersichtlich, dass $P(\,grau \mid Behälter_A\,) = \frac{2}{4}$ und $P(\,grau \mid Behälter_B\,) = \frac{1}{3}$ gilt.
Die bedingte Wahrscheinlichkeit kann formal auch folgendermaßen berechnet werden:

$$P(\,grau \mid Behälter_B\,) = \frac{P(\,grau\:und\:Behälter_B\,)}{P(\,Behälter_B\,)}$$<br />

Von besonderem Interesse ist der Satz von Bayes, der zugleich namensgebend für das betrachtete Verfahren ist. Dieser ermöglicht, bedingte Wahrscheinlichkeiten folgendermaßen zu manipulieren:

$$P(\,c \mid x\,) = \frac{P(\,c \mid x\,)\,\ast\,P(\,c\,)}{P(\,x\,)}$$

Auf diese Art lassen sich ohne Weiteres bedingte Wahrscheinlichkeiten invertieren.

### 1.3 Klassifizieren mit Hilfe von bedingten Wahrscheinlichkeiten

Bislang wurde die Qualifizierung nach der folgenden Regel vorgenommen:

Ist $P_1(x,y) > P_2(x,y)$, so ist der Datensatz der Klasse 1 zuzuordnen.<br />
Ist $P_2(x,y) > P_1(x,y)$, so ist der Datensatz der Klasse 2 zuzuordnen.

Dabei ist allerdings völlig unbekannt, worum es sich bei den Funktionen $P_1()$ und $P_2()$ jeweils handelt. Tatsächlich möchte man $P\,(c_1\mid x,y)$ und $P\,(c_2 \mid x,y)$ miteinander vergleichen. Dahinter verbirgt sich die folgende Fragestellung: „Gegeben ein Punkt $(x,y)$. Mit welcher Wahrscheinlichkeit ist er Klasse 1 zuzuordnen? Mit welcher Wahrscheinlichkeit ist er Klasse 2 zuzuordnen? Dabei ist für die Klassifizierung nicht der tatsächliche Wert entscheidend, sondern vielmehr welcher der Werte den größten Wert darstellt.

Unter Verwendung des bereits vorgestellten Satzes von Bayes, können die gesuchten Werte folgendermaßen berechnet werden:

$$P(\,c_i \mid x,y\,) = \frac{P(\,x,y \mid c_i\,)\,P(\,c_i\,)}{P(\,x, y\,)}$$

Mit Hilfe des Satz von Bayes ist es uns also möglich, diesen unbekannten Wert aus drei bekannten Werten zu errechnen, und damit eine Klassifizierungs-Prognose zu treffen, so dass die Klassifizierung schlussendlich nach der folgenden Regel vorgenommen werden kann:

Ist $P\,( c_1\mid x,y ) > P\,( c_2\mid x,y )$, so ist der Datensatz der Klasse $c_1$ zuzuordnen.<br />
Ist $P\,( c_1\mid x,y ) < P\,( c_2\mid x,y )$, so ist der Datensatz der Klasse $c_2$ zuzuordnen.



## 2 Unterschiedliche Ansätze der Wahrscheinlichkeitsberechnung mit Naïve Bayes

Der Naïve Bayes Klassifikator ist einer der effektivsten Algorithmus, welcher in Projekten für maschinelles Lernen implementiert wird. In erster Linie ist der Naïve Bayes ein linearer Klassifikator, der sowohl für die Kontrolle des maschinellen Lernens als auch für die Klassifizierung von Wahrscheinlichkeitsgruppen verwendet wird. 

<img src="bayes_classifer.png" width="50%" height="50%"/>

Der  Naïve Bayes Klassifikator nimmt an, dass die einzelnen Merkmale im Hinblick auf die Klassenvariable voneinander unabhängig sind. Dennoch haben die Merkmale eine gewissen Abhängigkeit. Bei einem Beispiel von Klassifizierung von Obst kann nach Form und Farbe gruppiert werden. Da Bananen nie rund sind, jedoch immer länglich sind, besitzt die Form eine Abhängigkeit zu der Frucht. Auch bei Wortfolgen können Abhängigkeiten zwischen den verschiedenen Worten beobachtet werden: so ist es in der deutschen Sprache beispielsweise nicht möglich, dass zwei Artikel unmittelbar aufeinander folgen. Unsere Sprachsyntax gibt dieses nicht her. Somit haben wir auch in diesem Beispiel eine Abhängigkeit. Da der Klassifikator von dieser Unabhängigkeit zwischen den Merkmalen ausgeht, ist die Berechnung schnell und effizient.

Im folgenden Kapitel werden drei Klassifikatoren vorgestellt, die für die Aufgabenstellung zur Verfügung stehen. Für die Implementierung wird nur ein Verfahren verwendet, weil es für den Einsatz von den Trainigs- und Testdaten der Buchstaben von Vorteil ist. Dieses wird im späteren Kapitel vorgestellt.

### 2.1 Bernoulli Verfahren 

Für den multivariaten Bernoulli Klassifikator werden vor allem Eingabegrößen mit unabhängigen boolschen Werten (binäre Variable) verwendet. Dieser Klassifikator wird gerne für die Klassifizierung von Dokumententexte, wie zum Beispiel die Verwendung von Spam-Filtern, wo eher nach bestimmten Wortvorkommnisse gesucht werden, als nach Häufigkeiten von  Wörter, verwendet.

$$P(\,x \mid C_k) = \prod_{i=1}^n P_{ki}^{x_i} (1 - P_{ki})^{(1-x_i)}$$

Wenn $x_i$ einen binären Wert darstellen soll, ob ein Wort vom i-ten Begriff des gesamten benutzten Vokabulars vorhanden oder nicht vorhanden ist, dann wird eine Klasse $C_k$ für die Wahrscheinlichkeit der Wortverwendung erstellt. $P_{ki}$ gibt die Wahrscheinlichkeit des Begriffs $x_i$ in der Klasse an. Häufig findet dieser Klassifikator Verwendung bei kurzen Texten.

### 2.2 Gauß'isches Verfahren

In der Gauß’ischen Normalverteilung wird die Handhabung von kontinuierlicher Verteilung, wie in unserem Beispiel die Testdaten der handgeschriebenen Buchstabe, berücksichtigt. So werden große Daten in einer Gauß’ischen Normalverteilung klassifiziert. 

Durch den Einsatz des Gauß’ischen Naïve Bayes Klassifikators vom SCIKIT-LEARN wird die Berechnung der Klassen mit einer hohen Trainingsgeschwindigkeit dargestellt. Als Ausgabe wird das Merkmal in einer Klasse zugeordnet, welches vorher durch die Eingabe der Trainingsdaten am wahrscheinlichsten entspricht. Bei Eingabe des Trainingsdaten werden vorher durch den Naïve Bayes Algorithmus die Klassen mit unterschiedlichsten Merkmalausprägungen erstellt. Wenn nun durch Eingabe der Testdaten diese Daten eingepflegt werden, wird durch den Gauß’ischen Naïve Bayes Klassifikator die Klassen erstellt. Der Klassifikator wird durch die Eingabe der Trainingsdaten realisiert. 

Wenn nun der Gauß’ische Naïve Bayes verwendet wird, werden kontinuierliche Daten normal verteilt. Mit der Annahme, dass die Trainingsdaten kontinuierlich steigern bzw. die Anzahl der Trainingsdaten sich stetig erhöht, wird eine Variable $x$ verwendet. Die Daten werden nun in eine bestimmte Anzahl an Klassen aufgeteilt und berechnen für jede Klasse $x$ den Mittelwert und die Varianz zur nächst benachbarten Klasse. Sei $\mu_c$ der Mittelwert jeder Klasse in $x$, wobei $c$ für jede Klasse gilt, und $\sigma_2$ die Varianz für jedes $x$ für jedes $c$. Weiterhin haben wir einen Beobachtungswert $v$. Somit wird eine Wahrscheinlichkeitsverteilung von $v$ für jedes $c$ und jeder Wahrscheinlichkeit von **$$P(x=v \mid c)=\frac{1}{\sqrt{2\pi\sigma^2_c}}\,e^{ -\frac{(v-\mu_c)^2}{2\sigma^2_c} }$$** <br />Bei Einfügen eines $v$ in die Gleichung einer Normalverteilung berechnet. Diese Verteilung wird vom Mittelwert und Varianz jeder Klasse parametrisiert. Im folgenden sieht man ein Beispiel einer grafischen Darstellung für die Klassifizierung mit dem Gauß’ische Naïve Bayes.

### 2.3 Multinomiales Verfahren

Die verschiedenen Klassifikatoren, die in Kombination mit dem naive Bayes verwendet werden können, sind entscheidend für die Berechnung der Wahrscheinlichkeit $P(\,(x,y)\,|\,C_i\,)$ also die Wahrscheinlichkeit, dass ein konkreter Featurevektor $(x,y)$ in der Klasse $C_i$ vorkommt.

$$p(\,x \mid C_k) = \frac{(\sum_i x_i)!}{\prod_i x_i !} \prod_i {p_{ki}}^{x_i}$$

Unter Berücksichtigung eines sogenannten "smoothing-Parameters" $\alpha$, der üblicherweise den Wert 1 hat, kann die genannte Formel auch folgendermaßen vereinfacht werden:

$$(1) \qquad P(\,x_k \mid c_j) = \frac{n_k + \alpha} {n + \alpha \cdot | Vocabulary |}$$
$$(2) \qquad P(c_j) = \frac{|docs_j|} {|total \, \# \, documents|}$$

Am Beispiel der Klassifikation von Texten verbergen sich dahinter die nachfolgenden Werten: $n_k$ stellt die Anzahl des Vorkommens des Wortes $x_k$ dar. Bei n handelt es sich um die Gesamtzahl der Worte in der Klasse $c_j$. $P(\,x_k \mid c_j)$ beschreibt dabei die bedingte Wahrscheinlichkeit, ein konkretes Wort $x_k$ in der Klasse $c_j$ zu finden. $P(c_j)$ wird auch als a-priori Wahrscheinlichkeit bezeichnet und gibt die Wahrscheinlichkeit an, mit der eine bestimmte Klasse $c_j$ erwartet wird.

Der multinomiale naive Bayes Klassifikator bietet sich insbesondere dann für die Klassifizierung an, wenn es sich bei den betrachteten Features um diskrete Wertebereiche handelt. Beispielsweise um die Anzahl von Worten in Dokumenten bei der Klassifizierung von Texten.

Benötigte Funktionsdefinitionen für die Matplot Grafiken (Learning Graph etc.), welche für das Darstellen der Lernkurve der unterschiedlichen Klassifikatoren benötigt werden.

```python
TEST_SIZE_PRECISION = 10
ALPHA_PRECISION = 3

char_distribution = [8.2, 1.5, 2.8, 4.3, 12.7, 2.2, 2.0, 6.1, 7.0, 0.2, 0.8, 4.0, 2.4, 6.7, 7.5, 1.9, 0.1, 6.0, 6.3,
                     9.1, 2.6, 1.0, 2.3, 0.2, 2.0, 0.1]
normalized_distribution = [float(i) / max(char_distribution) for i in char_distribution]

def plot_learning_curve(
    estimator,
    title,
    X,
    y,
    ylim=None,
    cv=None,
    n_jobs=1):
    """
    Generate a simple plot of the test and training learning curve.

    Parameters
    ----------
    estimator : object type that implements the "fit" and "predict" methods
        An object of that type which is cloned for each validation.

    title : string
        Title for the chart.

    X : array-like, shape (n_samples, n_features)
        Training vector, where n_samples is the number of samples and
        n_features is the number of features.

    y : array-like, shape (n_samples) or (n_samples, n_features), optional
        Target relative to X for classification or regression;
        None for unsupervised learning.

    ylim : tuple, shape (ymin, ymax), optional
        Defines minimum and maximum yvalues plotted.

    cv : int, cross-validation generator or an iterable, optional
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:
          - None, to use the default 3-fold cross-validation,
          - integer, to specify the number of folds.
          - An object to be used as a cross-validation generator.
          - An iterable yielding train/test splits.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:`StratifiedKFold` used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:`KFold` is used.

        Refer :ref:`User Guide <cross_validation>` for the various
        cross-validators that can be used here.

    n_jobs : integer, optional
        Number of jobs to run in parallel (default 1).
    """
    G_ALPHA = 0.1
    G_COLOR_1 = 'r'
    G_COLOR_2 = 'g'

    plt.title(title)

    if ylim is not None:
        plt.ylim(*ylim)

    plt.xlabel("Training examples")
    plt.ylabel("Score")

    train_sizes, train_scores, test_scores = learning_curve(
        estimator,
        X,
        y,
        cv=cv,
        exploit_incremental_learning=True,
        n_jobs=n_jobs,
        train_sizes=np.linspace(.9, .1, TEST_SIZE_PRECISION),
        verbose=0
    )
    
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)

    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid()
    plt.fill_between(
        train_sizes,
        train_scores_mean - train_scores_std,
        train_scores_mean + train_scores_std,
        alpha=G_ALPHA,
        color=G_COLOR_1
    )
    plt.fill_between(
        train_sizes,
        test_scores_mean - test_scores_std,
        test_scores_mean + test_scores_std,
        alpha=G_ALPHA,
        color=G_COLOR_2
    )

    plt.plot(
        train_sizes,
        train_scores_mean,
        'o-',
        color=G_COLOR_1,
        label="Training score"
    )
    plt.plot(
        train_sizes,
        test_scores_mean,
        'o-',
        color=G_COLOR_2,
        label="Cross-validation score"
    )
    
    plt.legend(loc="best")

    return plt

def init_plot(title, x_legend, y_legend):
    plt.title(title)

    plt.xlabel(x_legend)
    plt.ylabel(y_legend)

    plt.grid(True)


def plot(x, y, label):
    """Plot accuracy as a function of x."""
    x = np.array(x)
    y = np.array(y)

    plt.plot(x, y, label=label)


def print_plot():
    plt.legend()
    plt.show()
```

In dem folgenden Schritt war es die Aufgabe, die gegebenen Datensätze für das Maschinelle Lernen aufzuarbeiten. Ohne diese Buchstaben grafisch darzustellen haben wir die Matrizen normalisiert.

```python
def read_data(file_name):
    X = None
    y = None

    with open(file_name) as f:
        for l in f:
            # read data
            data = [l]
            for j in range(0, 13):
                data.append(f.readline().strip())
            data = list(map(float, " ".join(data).split()))

            X = np.array([data]) if X is None else np.append(X, [data], axis=0)

            y = np.array([f.readline().strip().split().index('0.80')]) if y is None else np.append(y, [
                f.readline().strip().split().index('0.80')], axis=0)

        y = np.array(y)

    return [X, y]
```

Verteilung und Normalisierung der gesplitteten Trainings- und Testsätze:

```python
def normalize_data(data):
    _data = np.array(data, copy=True)

    for l in np.nditer(_data, op_flags=['readwrite']):
        for e in np.nditer(l, op_flags=['readwrite']):
            e[...] = 0 if e == -0.5 else 1

    return _data


def get_distributed_subset(data, target):
    index_list = [
        [e for e in range(0, len(data), 26) if e == 1 or random.random() < normalized_distribution[0]],
        [e for e in range(1, len(data), 26) if e == 1 or random.random() < normalized_distribution[1]],
        [e for e in range(2, len(data), 26) if e == 2 or random.random() < normalized_distribution[2]],
        [e for e in range(3, len(data), 26) if e == 3 or random.random() < normalized_distribution[3]],
        [e for e in range(4, len(data), 26) if e == 4 or random.random() < normalized_distribution[4]],
        [e for e in range(5, len(data), 26) if e == 5 or random.random() < normalized_distribution[5]],
        [e for e in range(6, len(data), 26) if e == 6 or random.random() < normalized_distribution[6]],
        [e for e in range(7, len(data), 26) if e == 7 or random.random() < normalized_distribution[7]],
        [e for e in range(8, len(data), 26) if e == 8 or random.random() < normalized_distribution[8]],
        [e for e in range(9, len(data), 26) if e == 9 or random.random() < normalized_distribution[9]],
        [e for e in range(10, len(data), 26) if e == 10 or random.random() < normalized_distribution[10]],
        [e for e in range(11, len(data), 26) if e == 11 or random.random() < normalized_distribution[11]],
        [e for e in range(12, len(data), 26) if e == 12 or random.random() < normalized_distribution[12]],
        [e for e in range(13, len(data), 26) if e == 13 or random.random() < normalized_distribution[13]],
        [e for e in range(14, len(data), 26) if e == 14 or random.random() < normalized_distribution[14]],
        [e for e in range(15, len(data), 26) if e == 15 or random.random() < normalized_distribution[15]],
        [e for e in range(16, len(data), 26) if e == 16 or random.random() < normalized_distribution[16]],
        [e for e in range(17, len(data), 26) if e == 17 or random.random() < normalized_distribution[17]],
        [e for e in range(18, len(data), 26) if e == 18 or random.random() < normalized_distribution[18]],
        [e for e in range(19, len(data), 26) if e == 19 or random.random() < normalized_distribution[19]],
        [e for e in range(20, len(data), 26) if e == 20 or random.random() < normalized_distribution[20]],
        [e for e in range(21, len(data), 26) if e == 21 or random.random() < normalized_distribution[21]],
        [e for e in range(22, len(data), 26) if e == 22 or random.random() < normalized_distribution[22]],
        [e for e in range(23, len(data), 26) if e == 23 or random.random() < normalized_distribution[23]],
        [e for e in range(24, len(data), 26) if e == 24 or random.random() < normalized_distribution[24]],
        [e for e in range(25, len(data), 26) if e == 25 or random.random() < normalized_distribution[25]]
    ]
    index_list = [item for sublist in index_list for item in sublist]

    _data = [e for j, e in enumerate(data) if j in index_list]
    _target = [e for j, e in enumerate(target) if j in index_list]

    return np.array(_data), np.array(_target)


def my_train_test_split(X, y, test_size, random=True, is_relative_size=True):
    if is_relative_size:
        abs_test_size = round(test_size * len(X))
    else:
        abs_test_size = test_size
        test_size = abs_test_size / len(X)

    if random:
        head_X = X[:52]
        tail_X = X[52:]
        head_y = y[:52]
        tail_y = y[52:]

        X_train, X_test, y_train, y_test = train_test_split(tail_X, tail_y, test_size=test_size)

        X_train = np.append(X_train, head_X[26:], axis=0)
        y_train = np.append(y_train, head_y[26:], axis=0)

        X_test = np.append(X_test, head_X[:26], axis=0)
        y_test = np.append(y_test, head_y[:26], axis=0)
    else:
        X_train = np.array(X[abs_test_size:])
        X_test = np.array(X[:abs_test_size])
        y_train = np.array(y[abs_test_size:])
        y_test = np.array(y[:abs_test_size])

    return X_train, X_test, y_train, y_test


def get_priors(y):
    result = [0] * 26

    for e in y:
        result[e] += 1

    return [e / sum(result) for e in result]
```

### 3.2 Aufbau des Programms

Wir werden unsere Trainings- und Testdatensätze nun auslesen und diese trainieren wir mit den unterschiedlichen vorgestellten Klassifikatoren. Wir geben alle drei Klassifikatoren leer an. Das heißt, dass die Daten keine Erfolge im Training haben. Auch die Wahrscheinlichkeit, in wie weit die Buchstaben erkannt werden, liegt bei 0.

```python
X, y = read_data('data.pat')

best_gaussian = {
    'training_size': 0,
    'score': 0.0,
    }

best_bernoulli = {
    'training_size': 0,
    'score': 0.0,
    'alpha': 0.0,
    }

best_multinomial = {
    'training_size': 0,
    'score': 0.0,
    'alpha': 0.0,
    }
```

### 3.2.1 Gauß'ischer Klassifikator

```python
gaussian_accuracy = []
gaussian_training_duration = []
gaussian_scoring_duration = []
for test_size in np.linspace(.9, .1, TEST_SIZE_PRECISION):
    # init
    X_train, X_test, y_train, y_test = my_train_test_split(X, y, test_size=test_size)
    training_size = len(y_train)

    # classifier
    # cGauss = GaussianNB(priors=get_priors(y_train))
    cGauss = GaussianNB()

    # train
    t1 = time.time()
    cGauss.fit(X_train, y_train)
    training_duration = time.time() - t1

    # # predict
    # tmp = []
    # for e in X_test:
    #     t1 = time.time()
    #     cGauss.predict(e.reshape(1, -1))
    #     tmp.append(time.time() - t1)
    # prediction_duration = sum(tmp) / len(tmp)

    # score
    t1 = time.time()
    score = cGauss.score(X_test, y_test)
    scoring_duration = time.time() - t1
    if score > best_gaussian['score']:
        best_gaussian['score'] = score
        best_gaussian['training_size'] = training_size

    gaussian_accuracy.append((training_size, score))
    gaussian_training_duration.append((training_size, training_duration))
    gaussian_scoring_duration.append((training_size, scoring_duration))
```

### 3.2.2 Bernoulli Klassifikator

```python
bernoulli_accuracy = []
bernoulli_training_duration = []
bernoulli_scoring_duration = []
for alpha in np.linspace(.1, 1, ALPHA_PRECISION):
    for test_size in np.linspace(.9, .1, TEST_SIZE_PRECISION):
        # init
        X_train, X_test, y_train, y_test = my_train_test_split(X, y, test_size=test_size)
        training_size = len(y_train)

        # classifier
        cBernoulli = BernoulliNB(alpha=alpha, fit_prior=False, class_prior=get_priors(y_test))

        # train
        t1 = time.time()
        cBernoulli.fit(X_train, y_train)
        training_duration = time.time() - t1

        # # predict
        # tmp = []
        # for e in X_test:
        #     t1 = time.time()
        #     cBernoulli.predict(e.reshape(1, -1))
        #     tmp.append(time.time() - t1)
        # prediction_duration = sum(tmp) / len(tmp)

        # score
        t1 = time.time()
        score = cBernoulli.score(X_test, y_test)
        scoring_duration = time.time() - t1
        if score > best_bernoulli['score']:
            best_bernoulli['score'] = score
            best_bernoulli['alpha'] = alpha
            best_bernoulli['training_size'] = training_size

        bernoulli_accuracy.append((training_size, alpha, score))
        bernoulli_training_duration.append((training_size, alpha, training_duration))
        bernoulli_scoring_duration.append((training_size, alpha, scoring_duration))
```

### 3.2.3 Multinomial Klassifikator

```python
multinomial_accuracy = []
multinomial_training_duration = []
multinomial_scoring_duration = []
for alpha in np.linspace(.1, 1, ALPHA_PRECISION):
    for test_size in np.linspace(.9, .1, TEST_SIZE_PRECISION):
        # init
        X_train, X_test, y_train, y_test = my_train_test_split(X, y, test_size=test_size)
        training_size = len(y_train)

        # classifier
        cMultinomial = MultinomialNB(alpha=alpha, fit_prior=False, class_prior=get_priors(y_test))

        # train
        t1 = time.time()
        X_train = normalize_data(X_train)
        cMultinomial.fit(X_train, y_train)
        training_duration = time.time() - t1

        # # predict
        # tmp = []
        # for e in normalize_data(X_test):
        #     t1 = time.time()
        #     cMultinomial.predict(e.reshape(1, -1))
        #     tmp.append(time.time() - t1)
        # prediction_duration = sum(tmp) / len(tmp)

        # score
        t1 = time.time()
        score = cMultinomial.score(normalize_data(X_test), y_test)
        scoring_duration = time.time() - t1
        if score > best_multinomial['score']:
            best_multinomial['score'] = score
            best_multinomial['alpha'] = alpha
            best_multinomial['training_size'] = training_size

        multinomial_accuracy.append((training_size, alpha, score))
        multinomial_training_duration.append((training_size, alpha, training_duration))
        multinomial_scoring_duration.append((training_size, alpha, scoring_duration))
```

### Multinomial Klassifikator - Dynamik des A-priors-Faktor 

```python
cMultinomial = MultinomialNB(alpha=0.1, fit_prior=False, class_prior=[1 / 26] * 26)
data_p = []
data_np = []
for i in range(0, round(len(X) / 26) - 1):
    training_size_absolute = 26 * (i + 1)

    cMultinomialP = MultinomialNB(alpha=0.1, fit_prior=False, class_prior=normalized_distribution)
    cMultinomialNP = MultinomialNB(alpha=0.1, fit_prior=False, class_prior=[1 / 26] * 26)

    training_size_absolute = 26 * (i + 1)

    X_train = X[:training_size_absolute]
    X_test = X[training_size_absolute:]
    y_train = y[:training_size_absolute]
    y_test = y[training_size_absolute:]

    cMultinomialP.fit(normalize_data(X_train), y_train)
    cMultinomialNP.fit(normalize_data(X_train), y_train)

    X_test, y_test = get_distributed_subset(X_test, y_test)

    data_p.append((i + 1, cMultinomialP.score(normalize_data(X_test), y_test)))
    data_np.append((i + 1, cMultinomialNP.score(normalize_data(X_test), y_test)))

# debug
# print(data_p)
# print(data_np)
```

### Darstellung der verschiedenen Epochen beim Multinomial Klassifikator

```python
cMultinomial = MultinomialNB(alpha=0.1, fit_prior=False, class_prior=[1 / 26] * 26)
data_epo = []
for i in range(0, round(len(X) / 26) - 1):
    X_train = X[26 * i:][:26]
    X_test = X[26 * (i + 1):]
    y_train = y[26 * i:][:26]
    y_test = y[26 * (i + 1):]

    cMultinomial.partial_fit(normalize_data(X_train), y_train, classes=range(0, 26))

    data_epo.append((i + 1, cMultinomial.score(normalize_data(X_test), y_test)))

data_full = []
for i in range(0, round(len(X) / 26) - 1):
    cMultinomial = MultinomialNB(alpha=0.1, fit_prior=False, class_prior=[1 / 26] * 26)
    training_size_absolute = 26 * (i + 1)

    X_train = X[:training_size_absolute]
    X_test = X[training_size_absolute:]
    y_train = y[:training_size_absolute]
    y_test = y[training_size_absolute:]

    cMultinomial.fit(normalize_data(X_train), y_train)

    data_full.append((i + 1, cMultinomial.score(normalize_data(X_test), y_test)))
```

```python
# debug
# print(data_epo)
# print(data_full)

# Plot graphs
init_plot('Accuracy as a function of learned training sets', 'Training Sets', 'Accuracy')
plot([e[0] for e in data_epo], [e[1] for e in data_epo], 'Incremental Learning')
plot([e[0] for e in data_full], [e[1] for e in data_full], 'Full Learning')
print_plot()
```

### Falscherkennungsrate der Buchstaben durch den Multinominalen Klassifikator

```python
cMultinomial = MultinomialNB(alpha=0.1, fit_prior=False, class_prior=[1 / 26] * 26)

X_train = X[:520]
X_test = X[520:]
y_train = y[:520]
y_test = y[520:]

cMultinomial.fit(normalize_data(X_train), y_train)

results = {}
for _X, _y in zip(X_test, y_test):
    if not _y in results:
        results[_y] = [0] * 26
    results[_y][cMultinomial.predict(normalize_data(_X).reshape(1, -1))[0]] += 1

# print(results)

best = 0
worst = 26
for i, r in results.items():
    if r[i] < worst:
        worst = r[i]
        r_worst = {'char': chr(i + 97), 'r': r}
    if r[i] > best:
        best = r[i]
        r_best = {'char': chr(i + 97), 'r': r}

labels = [chr(i + 97) for i, e in enumerate(r_best['r']) if e > 0]
sizes = [e for i, e in enumerate(r_best['r']) if e > 0]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Best interpreted character "%s"' % r_best['char'])
ax1.axis('equal')
plt.show()

labels = [chr(i + 97) for i, e in enumerate(r_worst['r']) if e > 0]
sizes = [e for i, e in enumerate(r_worst['r']) if e > 0]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Worst interpreted character "%s"' % r_worst['char'])
ax1.axis('equal')
plt.show()
```
